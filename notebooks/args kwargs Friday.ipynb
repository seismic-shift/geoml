```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def args_func(*args):
    return args

def kwargs_func(**kwargs):
    return kwargs
```

```python
args_func(1, 2, 3, 4)
```

```python
kwargs_func(arg1='yellow', arg2=5, arg3=args_func)
```

```python
def both_func(*args, vp=2500, **kwargs):
    print(args)
    print(kwargs)
```

```python
both_func(1, 5, 6, sky='blue', ground='brown')
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.random(size=x.size) *0.1
```

```python
plt.plot(x, y, color='tab:orange', linewidth=2, linestyle='-.')
```

```python
plot_args = {'color': 'tab:orange', 'linewidth': 2, 'linestyle': '-.'}
```

```python
plt.plot(x, y, **plot_args)
```

```python
data = (x, y)
```

```python
plt.plot(*data, **plot_args)
```
